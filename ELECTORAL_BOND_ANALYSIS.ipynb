```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
buyer_data=pd.read_csv('electroral_bonds_buyer_data.csv',index_col=[0])
buyer_data.head()
```

```python
print("Companies that bought bonds :")
for i in buyer_data['Purchaser Name'].unique():
    print(i)
```

```python
print("Total amount of bonds purchased :",buyer_data['Denomination'].sum())
```

```python
# Set the style and font
sns.set(style="whitegrid")
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Georgia']
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# Calculate the amount of bonds bought each day
unique_dates = buyer_data['Date of Purchase'].unique()
amount_in_day = []
for i in unique_dates:
    amount_in_day.append(buyer_data[buyer_data['Date of Purchase'] == i]['Denomination'].sum() / 1000000)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(amount_in_day, marker='o', linestyle='-', color='royalblue', linewidth=2, markersize=6)

# Add title and labels with custom font size
plt.title('Daily Bond Purchase Amounts (in Millions INR)', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Index For the Day', fontsize=14, fontweight='bold')
plt.ylabel('Amount of Bond Bought/1000000 (INR)', fontsize=14, fontweight='bold')

# Customize grid and ticks
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
print(f"On {unique_dates[amount_in_day.index(max(amount_in_day))]} the max. amount of bounds were purchased : {max(amount_in_day) * 1000000} INR")
```

```python
Party_data=pd.read_csv('electroral_bonds_receiver_data.csv',index_col=[0])
Party_data.head()
```

```python
print("Parties that received bonds :")
for i in Party_data['Name of the Political Party'].unique():
    print(i)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data
party_funds = Party_data.groupby('Name of the Political Party')['Denomination'].sum().reset_index()
total_amount_received = party_funds['Denomination'].sum()
party_funds['Percentage'] = (party_funds['Denomination'] / total_amount_received) * 100
party_funds = party_funds.sort_values(by='Percentage', ascending=False)
party_pie = party_funds.head(8)

# Define a color palette and match it with party names
colors = sns.color_palette("husl", len(party_pie))
party_colors = dict(zip(party_pie['Name of the Political Party'], colors))

# Create pie chart
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(
    party_pie['Denomination'],
    labels=party_pie['Name of the Political Party'],
    autopct='%1.1f%%',
    startangle=140,
    colors=[party_colors[name] for name in party_pie['Name of the Political Party']],
    wedgeprops=dict(width=0.4, edgecolor='w'),
    textprops={'fontsize': 12, 'color': 'black'}
)

# Color-coordinate the labels with pie slices and set percentage text color to black
for text, autotext in zip(texts, autotexts):
    text.set_color(party_colors[text.get_text()])  # Color labels according to the pie slices
    autotext.set_color('black')  # Set percentage text color to black

plt.title('Percentage of Bonds Bought by Political Parties', fontsize=16, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set default font to a commonly available one
plt.rcParams['font.family'] = 'sans-serif'

# Example data loading
# Party_data = pd.read_csv('path_to_your_csv_file.csv')

# Group by 'Political Party' and sum 'Amount Received'
party_fundss = Party_data.groupby('Name of the Political Party')['Denomination'].sum().reset_index()

# Sort by 'Amount Received'
party_fundss = party_fundss.sort_values(by='Denomination', ascending=False)

# Convert 'Denomination' to crores
party_fundss['Received Amount'] = party_fundss['Denomination'] / 10000000

# Select top 20
top20 = party_fundss[['Name of the Political Party', 'Received Amount']].head(20)

# Define the number of colors
num_colors = len(top20)

# Generate a range of colors using a gradient
colors = plt.cm.inferno(np.linspace(0, 1, num_colors))

# Create horizontal bar chart
plt.figure(figsize=(13, 6))
barsss = plt.barh(top20['Name of the Political Party'], top20['Received Amount'], color=colors)

# Add labels for the bars
for bar, value in zip(barsss, top20['Received Amount']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{value:.2f}',
             va='center', ha='left', fontsize=12, color='black')

# Add titles and labels
plt.title("Top 20 Parties Receiving Funds from Electoral Bonds (in Crores)", fontsize=16)
plt.xlabel("Funds Received (in Crores)", fontsize=14)
plt.ylabel("Political Party", fontsize=14)

# Show the plot
plt.show()
```

```python
party_funds = Party_data.groupby(['Date of\nEncashment', 'Name of the Political Party'])['Denomination'].sum().reset_index()

# Convert 'Date of\nEncashment' to datetime
party_funds['Date of\nEncashment'] = pd.to_datetime(party_funds['Date of\nEncashment'])

# Extract the year from the 'Date of\nEncashment'
party_funds['Year'] = party_funds['Date of\nEncashment'].dt.year

# Calculate yearly funds received
yearly_funds_received = party_funds.groupby('Year')['Denomination'].sum().reset_index()

# Convert the amount into crores
yearly_funds_received['Denomination'] = yearly_funds_received['Denomination'] / 10000000

# Plotting the lineplot
plt.figure(figsize=(13, 6))
plt.title('Total Amount Received by Political Parties between Apr 12, 2019 to 11 Jan 2024')
sns.lineplot(x='Year', y='Denomination', data=yearly_funds_received, marker='o', color='b')

plt.ylabel('Amount Received in Crores')
plt.xlabel('Year')

# Show the plot
plt.show()
```

```python
# Set the style and font
sns.set(style="whitegrid")
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Georgia']
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# Convert 'Date of Encashment' to datetime
Party_data['Date of\nEncashment'] = pd.to_datetime(Party_data['Date of\nEncashment'], errors='coerce')

# Calculate the amount of bonds bought on each date
unique_dates = Party_data['Date of\nEncashment'].dropna().unique()
unique_dates = sorted(unique_dates)  # Convert to list and sort
amount_in_day = []
for date in unique_dates:
    amount_in_day.append(Party_data[Party_data['Date of\nEncashment'] == date]['Denomination'].sum() / 1000000)

# Create the plot
plt.figure(figsize=(14, 8))
plt.plot(unique_dates, amount_in_day, marker='o', linestyle='-', color='royalblue', markersize=8)

# Add title and labels with custom font size
plt.title('Bond Purchase Amounts Over Encashment Dates (in Millions INR)', fontsize=18, fontweight='bold', color='darkred')
plt.xlabel('Date of Encashment', fontsize=14, fontweight='bold')
plt.ylabel('Amount of Bond Bought/1000000 (INR)', fontsize=14, fontweight='bold')

# Format date on x-axis
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator())
plt.xticks(rotation=45)

# Customize grid and ticks
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```
